```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

- Abrir imagem:

```python
img = np.array(cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE))
rows, cols = img.shape

#img = cv2.copyMakeBorder(img, 0, cv2.getOptimalDFTSize(cols), 0, cv2.getOptimalDFTSize(rows), cv2.BORDER_CONSTANT, value=0)

plt.figure(1)
plt.imshow(img, 'gray')
plt.title("Imagem original")
plt.show()
```

```python
cv2.getOptimalDFTSize(cols)
```

- Aplicar a transformada discrea de Fourier:

```python
img_dft = np.fft.fft2(img)

#print(img_dft.shape)

#img_dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
```

- Trazer a componente DC para o centro da imagem, transladando n/2 nas duas direções: 

```python
img_dft_shift = np.fft.fftshift(img_dft)
```

- Calcular a magnitude a partir da componente real e imaginária:

```python
#img_dft_mag = (cv2.magnitude(img_dft_shift[:,:,0], img_dft_shift[:,:,1]))

img_dft_mag = np.abs(img_dft_shift)
```

```python
plt.figure(figsize=(10,9))
plt.subplot(121)
plt.imshow(img, 'gray')
plt.title("Imagem original")
plt.axis('OFF')

plt.subplot(122)
plt.imshow(20*np.log(img_dft_mag), 'gray')
plt.title("Espectro em frequência")
plt.axis('OFF')

plt.show()
```

- Cálculo da inversa:

```python
img_idft = np.fft.ifft2(img_dft)
img_inversa = np.abs(img_idft)
#print(img_idft)

plt.figure(1)
plt.imshow(img_inversa, 'gray')
plt.title("Imagem após IDFT")
plt.axis('OFF')

plt.show()
```

## Filtragem na frequência

### Filtro passa-baixa:

- Criar máscara gaussiana 

```python
def gaussianKernel(h1, h2):

    import numpy as np
    import matplotlib.pyplot as plt
    import math as m

    ## Returns a normalized 2D gauss kernel array for general purporses

    x, y = np.mgrid[0:h2, 0:h1]
    x = x-h2/2
    y = y-h1/2
    sigma = 1
    g = np.exp( -( x**2 + y**2 ) / (2*sigma**2) )
    return g / g.sum()

filterKernel = gaussianKernel(rows,cols)
#print(filterKernel)
```

```python
#filterKernel = cv2.copyMakeBorder(filterKernel, 0, cv2.getOptimalDFTSize(cols), 0, cv2.getOptimalDFTSize(rows), cv2.BORDER_CONSTANT, value=0)

filter_dft = np.fft.fft2(filterKernel)
filter_dft_shift = np.fft.fftshift(filter_dft)
filter_dft_mag = np.abs(filter_dft_shift)

#filter_dft = cv2.dft(np.float32(filterKernel),flags = cv2.DFT_COMPLEX_OUTPUT)
#filter_dft_shift = np.fft.fftshift(filter_dft)
#filter_dft_mag = 20*np.log(cv2.magnitude(filter_dft_shift[:,:,0], filter_dft_shift[:,:,1]))
#filter_dft_mag = (cv2.magnitude(filter_dft_shift[:,:,0], filter_dft_shift[:,:,1]))

plt.figure(1)
plt.imshow(filter_dft_mag, 'gray')
plt.title("Espectro em frequência do filtro Gaussiano com spread 1.5")
plt.show()
```

- Filtragem na frequência

```python
filter_img = img_dft_shift * filter_dft_shift

img_back = np.fft.fftshift(np.fft.ifft2(filter_img))

img_back_mag = np.abs(img_back)


plt.figure(1)
plt.imshow(img_back_mag, 'gray')
plt.title("Imagem após filtragem na frequência")
plt.show()
```
