### Definições preliminares

```python
import cv2 #OpenCV
import numpy as np #biblioteca para trablahar com arrays de forme eficiente
import matplotlib.pyplot as plt # biblioteca para plotar gráficos
```

## Carregar imagem

```python
img = cv2.imread('../figs/lena_color_256.tif', cv2.IMREAD_GRAYSCALE)
img
```

## Exibir imagem

```python
plt.figure()
plt.imshow(img, 'gray')
plt.show()
```

```python
# Image size
lin, col = np.shape(img)
print(lin)
print(col)
```

```python
# Pixels intensity
print('Max value: {}'.format(img.max()))
print('Min value: {}'.format(img.min()))
```

```python
# Create vector of pixels intensity
base_vector = np.zeros(256)
```

```python
for i in range(0,lin):
    for k in range(0,col):
        base_vector[img[i,k]] +=1
        
plt.plot(base_vector)
plt.show()
```

```python
# Using opencv
hist = cv2.calcHist([img], [0], None, [256], [0,256])
hist = hist.reshape((hist.shape[0],))
plt.plot(hist)
plt.show()
```

```python
# Using numpy
hist,bins = np.histogram(img.ravel(),256,[0,256])
plt.plot(hist)
plt.show()
```

```python
plt.stem(hist)
plt.show()
```
